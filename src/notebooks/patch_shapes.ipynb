```python
x.shape
```

```python
x
```

```python
patch_size = 2
```

```python
B, C, H, W = x.shape
P = patch_size
B, C, H, W, P
```

```python
# Crop to patch size multiplier as in PatchEmbed
x = x[:, :H // P * P, :W // P * P]
```

```python
x.shape
```

```python
x
```

`tensor.unfold(d, p, s)` extracts patches along dimention `d` with size `p` and stride `s`. The output is a tensor of shape `(*, n, ..., p)` where `n` is the new size of dimention `d` = the number of patches and new last dimention corresponds to each patch.

```python
x.unfold(2, P, P).shape
```

```python
x.unfold(2, P, P)
```

Applying `unfold` sequentially to `H` and `W` dimentions result in one before last dimention forresponding to patch `H` and last dimention for `W` (both sizes equal to `P` here). 

```python
x.unfold(2, P, P).unfold(3, P, P).shape
```

```python
x.unfold(2, P, P).unfold(3, P, P)
```

```python
# Extract patches
x = x \
    .unfold(2, P, P) \
    .unfold(3, P, P)
```

```python
x.shape
```

```python
x
```

We expect the first patch from the original to be the first patch in the unfolded tensor and second (in row order) being the second and so on.

```python
assert torch.allclose(
    x_orig[0, 0, :2, :2], 
    x[0, 0, 0, 0]
)
assert torch.allclose(
    x_orig[0, 0, :2, 2:4], 
    x[0, 0, 0, 1]
)
assert torch.allclose(
    x_orig[0, 0, 2:4, 2:4], 
    x[0, 0, 1, 1]
)
```

```python
# Concat to batch dimension
# (B, C, H_patches, W_patches, patch_size, patch_size) -> 
# (B * H_patches * W_patches, C, patch_size, patch_size)
H_patches, W_patches = x.shape[2:4]
H_patches, W_patches
```

```python
x.reshape(B * H_patches * W_patches, C, P, P).shape
```

```python
x = x.reshape(B * H_patches * W_patches, C, P, P)
```

```python
x.shape
```

```python
x
```

```python
H_patches, W_patches
```

```python
assert torch.allclose(
    x_orig[0, 0, :2, :2],
    x[0]
)
assert torch.allclose(
    x_orig[0, 0, :2, 2:4],
    x[1]
)
assert torch.allclose(
    x_orig[0, 0, :2, 4:6],
    x[2]
)
assert torch.allclose(
    x_orig[0, 0, :2, 6:8],
    x[3]
)
assert torch.allclose(
    x_orig[0, 0, 2:4, :2],
    x[4]
)
```
